```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

range_nr = 1000

df = pd.DataFrame(
    np.random.randn(range_nr, 4), # 10004
    index=pd.date_range('1/1/2000', periods=range_nr), # 
    columns=['A', 'B', 'C', 'D']
)

# cumsum DataFrame
df.cumsum().plot() # 
plt.title('Random Data Over Time')
plt.xlabel('Date')
plt.ylabel('Values')
plt.legend(loc="upper left")
plt.show()
```
