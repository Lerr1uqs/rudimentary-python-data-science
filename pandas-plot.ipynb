```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

range_nr = 1000

df = pd.DataFrame(
    np.random.randn(range_nr, 4), # 1000行4列矩阵
    index=pd.date_range('1/1/2000', periods=range_nr), # 行匹配
    columns=['A', 'B', 'C', 'D']
)

# cumsum 是对DataFrame进行累积求和的操作。具体来说，它沿着轴方向对数据进行累积求和，将每一行的值替换为该行及之前所有行的和。
df.cumsum().plot() # 默认应该是
plt.title('Random Data Over Time')
plt.xlabel('Date')
plt.ylabel('Values')
plt.legend(loc="upper left")
plt.show()
```
